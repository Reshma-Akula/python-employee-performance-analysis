Task 1:
There are three different tables as given above. Please make three dataframe in python
and save them as three .csv files.

```python
import pandas as pd

# -------------------------------
# 1. Project DataFrame
# -------------------------------

# Creating dictionary for project data
project_data = {
    "ID": ["A001","A002","A003","A004","A005","A002","A005","A003","A001","A003","A001","A004","A004","A005"],
    
    # Project names
    "Project": ["Project 1","Project 2","Project 3","Project 4","Project 5","Project 6","Project 7","Project 8",
                "Project 9","Project 10","Project 11","Project 12","Project 13","Project 14"],
    
    # Cost of each project (None = missing value)
    "Cost": [1002000,2000000,4500000,5500000,None,680000,400000,350000,None,300000,2000000,1000000,3000000,200000],
    
    # Status of project
    "Status": ["Finished","Ongoing","Finished","Ongoing","Finished","Failed","Finished","Failed",
               "Ongoing","Finished","Failed","Ongoing","Finished","Finished"]
}


# Convert dictionary into DataFrame
df_project = pd.DataFrame(project_data)


# -------------------------------
# 2. Employee DataFrame
# -------------------------------

# Creating dictionary for employee details
employee_data = {
    "ID": ["A001","A002","A003","A004","A005"],  # Employee IDs
    
    "Name": ["John Alter","Alice Luxumberg","Tom Sabestine","Nina Adgra","Amy Johny"],  # Names
    
    "Gender": ["M","F","M","F","F"],  # Gender
    
    "City": ["Paris","London","Berlin","Newyork","Madrid"],  # City
    
    "Age": [25,27,29,31,30]  # Age of employees
}

# Convert dictionary into DataFrame
df_employee = pd.DataFrame(employee_data)


# -------------------------------
# 3. Seniority Level DataFrame
# -------------------------------

# Creating dictionary for designation levels
seniority_data = {
    "ID": ["A001","A002","A003","A004","A005"],  # Employee IDs
    
    "Designation Level": [2,2,3,2,3]  # Seniority level
}

# Convert dictionary into DataFrame
df_seniority = pd.DataFrame(seniority_data)


# -------------------------------
# 4. Save DataFrames as CSV files
# -------------------------------

# Save project data into CSV
df_project.to_csv("project_data.csv", index=False)

# Save employee data into CSV
df_employee.to_csv("employee_data.csv", index=False)

# Save seniority data into CSV
df_seniority.to_csv("seniority_data.csv", index=False)
```

Task 2:
The cost column in the dataframe “Project” has some missing values. Your task is to
compute these missing values. Replace the missing values by running average. You
should use the “For” loop for this task.

```python
import pandas as pd
import numpy as np

# Initialize variables to keep track of running total and count
running_sum = 0   # stores sum of all previous non-missing cost values
count = 0         # stores count of non-missing values

# Loop through each row in the dataframe
for i in range(len(df_project)):
    
    # Check if current 'cost' value is NOT missing (i.e., not NaN)
    if not pd.isna(df_project.loc[i, 'Cost']):
        
        # Add the current value to running sum
        running_sum += df_project.loc[i, 'Cost']
        
        # Increase count of valid values
        count += 1
    
    else:
        # If value is missing (NaN), replace it with running average
        
        if count > 0:
            # Compute running average = sum of previous values / count
            df_project.loc[i, 'Cost'] = running_sum / count
        else:
            # If first value itself is NaN (no previous values exist)
            df_project.loc[i, 'Cost'] = 0  # you can change this if needed

# Display updated dataframe
pd.options.display.float_format = '{:.0f}'.format
print(df_project)
df_project['Cost'] = df_project['Cost'].round(1)

# Creating new .csv file to fill missing values 
#df_project.to_csv("updated_project_data.csv", index=False)  # Creating new .csv file to fill missing values 

# Updating .csv file to fill missing values 
df_project.to_csv("project_data.csv", index=False)
```

Task 3:
Split the name column in the Employee dataframe into two new columns “First Name”,
and “LastName” and remove the older “name” column.

```python
# Load fresh employee data from CSV file (reset data every time)
df_employee = pd.read_csv("employee_data.csv")

# Split 'name' column into two columns
df_employee[['First Name', 'LastName']] = df_employee['Name'].str.split(' ', expand=True)

# Drop the original 'name' column
df_employee.drop('Name', axis=1, inplace=True)

# Display result
print(df_employee.head())
```

Task 4:
Join all three dataframes in one single dataframe. Name it “Final”

```python
# Step 1: Merge employee and project data
df_temp = pd.merge(df_employee, df_project, on='ID')

# Step 2: Merge with seniority data
Final = pd.merge(df_temp, df_seniority, on='ID')

#Increasing the display width
pd.set_option('display.width', 1000)

# Step 3: Display final dataframe
print(Final)

#creating a copy of final for future usage
Final_master = Final.copy()
```

Task 5:
Add a new bonus column in the Final dataframe. Give a 5% bonus concerning project 
cost only to employees who have finished the projects.

```python
# Create Bonus column with condition
Final['Bonus'] = Final.apply(
    lambda x: x['Cost'] * 0.05 if x['Status'] == 'Finished' else 0,
    axis=1
)

# Display result
print(Final)
```

Task 6:
Demote the designation level by 1, whose projects have status “fail”. Delete the
employees record whose designation level is above 4

```python
# Load fresh final data
Final = Final_master.copy()

# Demote level for failed projects
Final.loc[Final['Status'] == 'Failed', 'Designation Level'] = Final['Designation Level'] - 1

# Delete records where Level > 4
Final = Final[Final['Designation Level'] <= 4]

# Display result
print(Final)
```

Task 7
Add “Mr.” and “Mrs.” to the first name column and drop the gender column.

```python
# Load fresh final data
Final = Final_master.copy()

# Add prefix based on Gender
Final['First Name'] = Final.apply(
    lambda x: ('Mr. ' + x['First Name']) if x['Gender'] == 'M' 
    else ('Mrs. ' + x['First Name']),
    axis=1
)

# Drop Gender column
Final.drop('Gender', axis=1, inplace=True)

# Display result
print(Final)
```

Task 8:
Promote designation level by 1 for the employees whose age is more than 29 years using
IF condition

```python
# Load fresh final data
Final = Final_master.copy()

# Promote level for employees with Age > 29
Final['Designation Level'] = Final.apply(
    lambda x: x['Designation Level'] + 1 if x['Age'] > 29 else x['Designation Level'],
    axis=1
)

# Display result
print(Final)
```

Task 9:
Add the cost of all projects for each Employee and save it in new dataframe
“TotalProjCost” with three columns ID, First Name, and Total cost

```python

# Step 1: Group the project_data by 'ID' and calculate total 'Cost' for each employee
total_cost = df_project.groupby('ID')['Cost'].sum().reset_index()

# Step 2: Merge the total_cost dataframe with employee_data to get 'First Name'
# We are selecting only 'ID' and 'First Name' columns from employee_data
TotalProjCost = pd.merge(total_cost, df_employee[['ID', 'First Name']], on='ID')

# Step 3: Rearrange the columns in the required order: ID, First Name, Total cost
TotalProjCost = TotalProjCost[['ID', 'First Name', 'Cost']]

# Step 4: Rename the 'Cost' column to 'Total cost' as per requirement
TotalProjCost.rename(columns={'Cost': 'Total cost'}, inplace=True)

# Step 5: Display the final dataframe
print(TotalProjCost)
```

Task 10:
Print all the employee details whose city name contains the letter “o” in it

```python
# Step 1: Use str.contains() to filter rows where 'City' contains letter 'o'
# case=False → makes it case-insensitive (O or o both will match)
# na=False → ignores NaN values (avoids errors)
filtered_data = df_employee[df_employee['City'].str.contains('o', case=False, na=False)]

# Step 2: Print the filtered employee details
print(filtered_data)
```
